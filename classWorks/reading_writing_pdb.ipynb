## Working with PDB Files

```python
#!/usr/bin/env python3

import sys
import matplotlib.pyplot as plt
```

```python
pdbname=sys.argv[1]
```

```python
f=open("2FA9noend.pdb",'r')
lines=f.readlines()
```

```python
tmplist=[]
for line in lines:
    words=line.split()
    tempfact=float(words[10])
    #print (tempfact,line[60:66])
    tmplist.append(tempfact)
    #sys.exit()
f.close()
```

```python
f=open("tmp.out",'w')
for temp in tmplist:
    s="The temperature factor is {0:6.3f}\n"
    f.write(s.format(temp))
f.close()
```

```python
plt.plot(tmplist)
plt.show()

print ("Done!")
```
