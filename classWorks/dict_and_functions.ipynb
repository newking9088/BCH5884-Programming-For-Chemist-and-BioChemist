```python
import sys
#pdbFileName = sys.argv[1]
pdbFileName = input("Please enter the file name: ")
```

```python
with open(pdbFileName) as f:
    readFile = f.readlines()
```

```python
records = []
massDict ={"H":1.01, "C":12.01, "N":14.01, "O": 16.0, "P": 30.97, "S":32.07, "MG": 24.30}
for line in readFile:
    atomDict = {}
    x = float(line[30:38])
    atomDict['x'] = x
    element = line[76:78].strip()
    atomDict['element'] = element
    mass = massDict[element]
    atomDict['mass'] = massDict[element]
    records.append([x,element,mass])
    records.append(atomDict)
records;
```

```python
# for importing purpose, name this as intersection.py
if __name__ =="__main__":
    import random
    def intersection(seq1, seq2):
        inBoth = []
        for n in seq1:
            if n in seq2:
                if n not in inBoth:
                    inBoth.append(n)
        return inBoth
```

```python
intersection([1,2,3,4],[3,4,5,6])
```

```python
nelements = 10
randint1 = []
randint2 = []
for n in range(nelements):
    randint1.append(random.randint(1,5))
    randint2.append(random.randint(1,5))
commonElements = sorted(intersection(randint1,randint1))
commonElements
```
